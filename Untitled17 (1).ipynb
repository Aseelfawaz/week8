```python
from datasets import load_dataset
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from transformers import AutoTokenizer, AutoModelForMaskedLM, Trainer, TrainingArguments
```

# Load dataset


```python
ds = load_dataset("CUTD/sanad_df")
df = pd.DataFrame(ds['train'])
```

```python
df
```

```python
import matplotlib.pyplot as plt
df['label'].value_counts().plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
```

# Encoding the labels


```python
encoder = LabelEncoder()
df['label'] = encoder.fit_transform(df['label'])
```

#spliting the data

```python
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv02")
model = AutoModelForMaskedLM.from_pretrained("aubmindlab/bert-base-arabertv02")
```

# Preprocess the data


```python
def preprocess_data(sentences):
    inputs = tokenizer(sentences['text'], truncation=True, padding='max_length', max_length=512, return_tensors='pt')
    inputs['labels'] = inputs['input_ids'].clone()#match re input id

    if isinstance(sentences['label'][0], str):  # Check if labels are strings
        unique_labels = set(sentences['label'])
        label_map = {label: i for i, label in enumerate(unique_labels)}
        inputs['label'] = [label_map[l] for l in sentences['label']]
    else:
        inputs['label'] = sentences['label']
    return inputs
```

```python
tokenizer_data = ds.map(preprocess_data, batched=True, remove_columns=["Unnamed: 0"])
```

#training arguments


```python
training_args = TrainingArguments(
    output_dir='./results',
    overwrite_output_dir=True,
    evaluation_strategy="no",
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    num_train_epochs=1
)

```

# Initialize Trainer


```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenizer_data['train'],
)
```

# Start training


```python
trainer.train()
```

#Inference

```python
from transformers import pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model="aubmindlab/bert-base-arabertv02") #يسهل اعلى مودل يسوي تاسك معين
```
